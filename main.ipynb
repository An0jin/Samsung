```python
import pandas as pd
df = pd.read_excel('./SamSung.xlsx')
df
```

```python
df['전일비']=df['전일비'].astype(str)
df.info()
```

```python
df['전일비']=df['전일비'].apply(lambda x: x[:int(len(x)/2)])
df
```

```python
# 각 칼럼별 산포도 그래프
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(font="NanumGothic", rc={"axes.unicode_minus": False})

sns.pairplot(df)
plt.show()
```

```python
import matplotlib.pyplot as plt

numerical_cols = df.select_dtypes(include=[int, float]).columns

fig, axes = plt.subplots(nrows=len(numerical_cols), ncols=1, figsize=(20, 20), sharex=True)

for i, col in enumerate(numerical_cols):
    df[col].plot(ax=axes[i], marker='o')
    axes[i].set_title(col)

plt.tight_layout()
plt.show()
plt.savefig('lineplt.png')
```
